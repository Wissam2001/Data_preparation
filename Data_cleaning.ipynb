```python
#Question 20: Plot the price changing

Vis_tp = df.copy()
```

```python
#here w're going to plot the price changing for each commodity
## so first of all we visualize the comodity we have in this dataframe
c= Vis_tp['commodity'].unique()
c
```

```python
#ploting the price changing of eatch commodity
for i in range(len(c)):
    plt.rcParams["figure.figsize"]=(7,5)
    plt.title(c[i])
    plt.ylabel('Price')
    
    #we're gonna to extruct the dataframe to ph whene it realise the condition
    ph=Vis_tp[Vis_tp['commodity'] ==c[i]]
    
    plt.plot(ph['price'])
    plt.show()
```

```python
#ploting the price changing of eatch commodity
for i in range(len(c)):
    plt.rcParams["figure.figsize"]=(7,5)
    plt.title(c[i])
    plt.ylabel('Price')
    
    #we're gonna to extruct the dataframe to ph whene it realise the condition
    ph=Vis_tp[Vis_tp['commodity'] ==c[i]]
    plt.subplot(10,10,i+1)
    plt.plot(ph['price'])
    plt.show()
```
